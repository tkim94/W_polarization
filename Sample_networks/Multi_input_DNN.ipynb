# Importing Necessary Packages

```python
import csv
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np
import pylab as pl
import seaborn as sns
import sklearn

from keras.models import Sequential,Model
from keras.models import Model
from keras.layers import Input
from keras.layers import Flatten
from keras.layers import Dropout
from keras.layers import Embedding
from keras.layers import Dense, Activation
from keras.layers import Dropout, BatchNormalization
from keras.callbacks import ReduceLROnPlateau,EarlyStopping
from matplotlib import pyplot
from keras.layers.convolutional import Conv2D, MaxPooling2D
from keras.layers.merge import concatenate
```

## Neutral Network Building

```python
from keras.utils import plot_model
import pydot, graphviz

# Channel 1
jet_2in = Input(shape=(28, 28, 1))
jet_2 = Conv2D(64, kernel_size=4, activation = 'relu')(jet_2in)
jet_2 = Dropout(0.5)(jet_2)
jet_2 = MaxPooling2D(pool_size=(2, 2), strides=(2, 2))(jet_2)
jet_2 = Conv2D(70, kernel_size=3, padding="same", activation = 'relu')(jet_2)
jet_2 = MaxPooling2D(pool_size=(2, 2), strides=(2, 2))(jet_2)
jet_2 = Flatten()(jet_2)

dense_5 = Dense(units=100, activation='relu')(jet_2)

# Channel 2
jet_1in = Input(shape=(28, 28, 1))
jet_1 = Conv2D(40, kernel_size=5, activation = 'relu')(jet_1in)
jet_1 = Dropout(0.5)(jet_1)
jet_1 = MaxPooling2D(pool_size=(2, 2), strides=(2, 2))(jet_1)
jet_1 = Conv2D(70, kernel_size=3, padding="same", activation = 'relu')(jet_1)
jet_1 = MaxPooling2D(pool_size=(2, 2), strides=(2, 2))(jet_1)
jet_1 = Flatten()(jet_1)

dense_4 = Dense(units=100, activation='relu')(jet_1)

# Merge channel 1 & 2
concatenate_2 = concatenate([dense_5,dense_4])

dense_6 = Dense(units=100, activation = 'relu')(concatenate_2)

# Channel 3
lep_1in = Input(shape=(28, 28, 1))
lep_1 = Conv2D(64, kernel_size=4, activation = 'relu')(lep_1in)
lep_1 = Dropout(0.5)(lep_1)
lep_1 = MaxPooling2D(pool_size=(2, 2), strides=(2, 2))(lep_1)
lep_1 = Conv2D(70, kernel_size=3, padding="same", activation = 'relu')(lep_1)
lep_1 = MaxPooling2D(pool_size=(2, 2), strides=(2, 2))(lep_1)
lep_1 = Flatten()(lep_1)

dense_1 = Dense(units=100, activation='relu')(lep_1)

# Channel 4
lep_2in = Input(shape=(28, 28, 1))
lep_2 = Conv2D(40, kernel_size=4, activation = 'relu')(lep_2in)
lep_2 = Dropout(0.5)(lep_2)
lep_2 = MaxPooling2D(pool_size=(2, 2), strides=(2, 2))(lep_2)
lep_2 = Conv2D(70, kernel_size=3, padding="same", activation = 'relu')(lep_2)
lep_2 = MaxPooling2D(pool_size=(2, 2), strides=(2, 2))(lep_2)
lep_2 = Flatten()(lep_2)

dense_2 = Dense(units=100, activation='relu')(lep_2)

# Merge channel 1 & 2
concatenate_1 = concatenate([dense_1,dense_2])

dense_3 = Dense(units=100, activation = 'relu')(concatenate_1)

# Channel 5
MET = Input(shape=(10,))
dense_7 = Dense(units=100, activation = 'relu')(MET)

# Merge dense 6, 3 & 7
concatenate_3 = concatenate([dense_6, dense_3, dense_7])

dense_8 = Dense(units=100, activation = 'relu')(concatenate_3)
dense_9 = Dense(units=1, activation = 'tanh')(dense_8)

model = Model(inputs=[jet_2in, jet_1in, lep_1in, lep_2in, MET], outputs=dense_9)
#model.summary()

plot_model(model, show_shapes=True, show_layer_names=True,to_file='multi_input_DNN.pdf')
```
